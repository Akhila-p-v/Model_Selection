**HANDLING MISSING VALUES**

****

```python
print(df.isnull().sum()) # check if there are any remaining null values in the dataset before filling missing values
```

```python
from sklearn.impute import SimpleImputer
# Impute missing values in the 'Age' column with the median age
imputer = SimpleImputer(strategy='median')
df['Age'] = imputer.fit_transform(df[['Age']])
# Fill missing values in the 'Embarked' column with the mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
# Drop non-numeric and irrelevant columns
df.drop(columns=['Name', 'Ticket', 'Cabin'], inplace=True)
```

```python
print(df.isnull().sum()) # check if there are any remaining null values in the dataset after filling missing values
```

**ONE HOT ENCODING**

```python
from sklearn.preprocessing import OneHotEncoder
# One-hot encode the 'Sex' and 'Embarked' columns
encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(df[['Sex', 'Embarked']]).toarray()

# Get the feature names from the encoder
sex_categories = encoder.categories_[0]
embarked_categories = encoder.categories_[1]

# Create DataFrame with encoded features
encoded_df = pd.DataFrame(encoded_features, columns=list(sex_categories) + list(embarked_categories))

# Concatenate the encoded columns with the original DataFrame
df = pd.concat([df, encoded_df], axis=1)

# Drop the original categorical columns ('Sex', 'Embarked')
df.drop(columns=['Sex', 'Embarked'], inplace=True)

```

**k-fold and stratified k-fold cross-validation**

```python
from sklearn.model_selection import cross_val_score, KFold, StratifiedKFold
```

```python

# Define features (X) and target variable (y)
X = df.drop(columns=['Survived'])  # Features excluding 'Survived'
y = df['Survived']  # Target variable
```

```python
# Define the number of folds for cross-validation
num_folds = 5
```

```python
# Perform k-fold cross-validation for kNN
kfold = KFold(n_splits=num_folds, shuffle=True, random_state=42)
knn_scores = cross_val_score(knn_model, X, y, cv=kfold)
```

```python
# Perform stratified k-fold cross-validation for kNN
stratified_kfold = StratifiedKFold(n_splits=num_folds, shuffle=True, random_state=42)
stratified_knn_scores = cross_val_score(knn_model, X, y, cv=stratified_kfold)
```

```python
# Perform k-fold cross-validation for SVM
svm_scores = cross_val_score(svm_model, X, y, cv=kfold)

# Perform stratified k-fold cross-validation for SVM
stratified_svm_scores = cross_val_score(svm_model, X, y, cv=stratified_kfold)
```

```python
# Calculate average accuracy scores
knn_avg_accuracy = knn_scores.mean()
svm_avg_accuracy = svm_scores.mean()
stratified_knn_avg_accuracy = stratified_knn_scores.mean()
stratified_svm_avg_accuracy = stratified_svm_scores.mean()

# Print average accuracy scores
print("Average Accuracy Scores (kNN - k-fold):", knn_avg_accuracy)
print("Average Accuracy Scores (kNN - Stratified k-fold):", stratified_knn_avg_accuracy)
print("Average Accuracy Scores (SVM - k-fold):", svm_avg_accuracy)
print("Average Accuracy Scores (SVM - Stratified k-fold):", stratified_svm_avg_accuracy)
```

** SVM appears to perform slightly better than kNN in this case based accuracy**

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix

# Fit the models
knn_model.fit(X, y)
svm_model.fit(X, y)
```

```python
# Predictions
knn_preds = knn_model.predict(X)
svm_preds = svm_model.predict(X)
```

```python

# Calculate metrics for kNN
knn_precision = precision_score(y, knn_preds)
knn_recall = recall_score(y, knn_preds)
knn_f1 = f1_score(y, knn_preds)
knn_conf_matrix = confusion_matrix(y, knn_preds)

# Calculate metrics for SVM
svm_precision = precision_score(y, svm_preds)
svm_recall = recall_score(y, svm_preds)
svm_f1 = f1_score(y, svm_preds)
svm_conf_matrix = confusion_matrix(y, svm_preds)

# Print metrics for kNN
print("Metrics for kNN:")
print("Precision:", knn_precision)
print("Recall:", knn_recall)
print("F1-score:", knn_f1)
print("Confusion Matrix:")
print(knn_conf_matrix)

# Print metrics for SVM
print("\nMetrics for SVM:")
print("Precision:", svm_precision)
print("Recall:", svm_recall)
print("F1-score:", svm_f1)
print("Confusion Matrix:")
print(svm_conf_matrix)
```

SVM model achieved higher precision but then it suffered from lower recall compared to the kNN model.
